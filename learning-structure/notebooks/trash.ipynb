```python
import torch
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = torch.randn((1000, 3), requires_grad=False)
```

```python
row_norms = torch.linalg.vector_norm(x, dim=1)
row_norms.shape
```

```python
y = x / row_norms.reshape(x.shape[0], 1)
torch.linalg.vector_norm(y, dim=1)
```

```python
show_grid = True
show_ticks = True

fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig)

ax.scatter(y[:, 0].numpy(), y[:, 1].numpy(), y[:, 2].numpy())

ax.grid(show_grid)

ax._axis3don = show_ticks
    
plt.show()
```
